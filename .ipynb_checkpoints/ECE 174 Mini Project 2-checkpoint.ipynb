## Functions

### Neural Network Function
f(x,w) - function takes a vector of inputs and vector of weight as inputs and outputs single number  
function(x,w) - Uses the 'f' function to populate an array with the result of every input value in the data with a given vector of weights  


```python
def f(x, w):
    return (w[0]*np.tanh(x[0])* (w[1]*x[0] + w[2]*x[1] + w[3]*x[2] + w[4]*1)
            + (w[5]*np.tanh(x[1])* (w[6]*x[0] + w[7]*x[1] + w[8]*x[2] + w[9]*1))
            + (w[10]*np.tanh(x[2])* (w[11]*x[0] + w[12]*x[1] + w[13]*x[2] + w[14]*1) + w[15]))
```

```python
def function(x, w, N):
    fx = np.empty((N,1))
    for i in range(N):
        funct = f(x[i], w)
        fx[i] = funct

    return fx
```

### Partial Derivatives 

```python
def dfw1(x,w):
    return 1-(np.tanh(x[0]) * np.tanh(x[0]))
def dfw2(x,w):
    return x[0]
def dfw3(x,w):
    return x[1]
def dfw4(x,w):
    return x[2]
def dfw5(x,w):
    return 1

def dfw6(x,w):
    return 1-(np.tanh(x[1]) * np.tanh(x[1]))
def dfw7(x,w):
    return x[0]
def dfw8(x,w):
    return x[1]
def dfw9(x,w):
    return x[2]
def dfw10(x,w):
    return 1

def dfw11(x,w):
    return 1-(np.tanh(x[2]) * np.tanh(x[2]))
def dfw12(x,w):
    return x[0]
def dfw13(x,w):
    return x[1]
def dfw14(x,w):
    return x[2]
def dfw15(x,w):
    return 1

def dfw16(x,w):
    return 1
```

## Jacobian Matrix using Partial Derivative Functions

```python
def deriv(x, w, N):

    D = np.empty((N, 16))

    for i in range(N):
        D[i, 0] = dfw1(x[i], w)
        D[i, 1] = dfw2(x[i], w)
        D[i, 2] = dfw3(x[i], w)
        D[i, 3] = dfw4(x[i], w)
        D[i, 4] = dfw5(x[i], w)
        D[i, 5] = dfw6(x[i], w)
        D[i, 6] = dfw7(x[i], w)
        D[i, 7] = dfw8(x[i], w)
        D[i, 8] = dfw9(x[i], w)
        D[i, 9] = dfw10(x[i], w)
        D[i, 10] = dfw11(x[i], w)
        D[i, 11] = dfw12(x[i], w)
        D[i, 12] = dfw13(x[i], w)
        D[i, 13] = dfw14(x[i], w)
        D[i, 14] = dfw15(x[i], w)
        D[i, 15] = dfw16(x[i], w)
        
    return D
```

### Sum of the Squared Residuals

```python
def square_sum_resid(x, y, N):
    for i in range(N):
        R = 0
        r = y[i] - x[i]
        R += (r*r)
    return R[0]
```

## Levenberg-Marquardt Algorithm

### Termination Conditions

The termination condition is set  to 0.01 since it is close enough to zero, and it allows the algorithm to be run in a timely manner. The max number of iterations is set to 25 to prevent crashing in the case the algoirthm does not converge

```python
termination = 0.005
max_iterations = 24
```

### Algorithm

```python
def LM(x, y, w, lamb, N):
    i = 0
    k2 = function(x, w, N)
    r2 = square_sum_resid(k2, y, N)
    R = np.array([])
    R = np.append(R, r2).astype(float)
    w_1 = w
    
    while(r2 > termination):
        D = deriv(x, w, N)
        b = D@w - function(x, w, N)
        I = np.identity(16)

        w_1 = (np.linalg.pinv(D.T@D+(lamb*(w.T@w)))@(D.T@b + lamb*(I.T@w)))

        k1 = function(x, w, N)
        k2 = function(x, w_1, N)

        r1 = square_sum_resid(k1, y, N)
        r2 = square_sum_resid(k2, y, N)
        
        R = np.append(R, r2).astype(float)
        
        if (r2 < r1):
            lamb = 0.8*lamb
            w = w_1
        else:
            lamb = 2*lamb
            w = w

        i+=1
        
        if (i == max_iterations):
            break
        
    iterations = np.array(list(range(0, i+1)))
    
    return iterations, R, w_1
    
    
```

## Training the Model

```python
lamb = 0.00001
iterations, residual, weights = LM(data, y, w_init, lamb, 500)
plt.figure(figsize=(12, 8))
plt.plot(iterations, residual, '--bo')
plt.axhline(y=termination, color='r', linestyle='-')
```

# Levenberg-Marquardt with Sigmoid Function

In this part, most of the prior code will be reused, but the fucntions that are used in the LM functions need to rewritten for the use of the sigmoid function instead of tanh. Once the functions are rewritten, the LM function will need to be rewritten to use the new "sigmoid" functions.
There is no need to generate new random data points. For the sake of clarity, we will call the sigmoid function g(x)

## Rewriting the Functions

### Functions (Sigmoid)

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
def g(x, w):
    return (w[0]*sigmoid(x[0])* (w[1]*x[0] + w[2]*x[1] + w[3]*x[2] + w[4]*1)
            + (w[5]*sigmoid(x[1])* (w[6]*x[0] + w[7]*x[1] + w[8]*x[2] + w[9]*1))
            + (w[10]*sigmoid(x[2])* (w[11]*x[0] + w[12]*x[1] + w[13]*x[2] + w[14]*1) + w[15]))
```

```python
def gfunction(x, w, N):
    gx = np.empty((N,1))
    for i in range(N):
        gfunct = g(x[i], w)
        gx[i] = gfunct

    return gx
```

### Partial Derivatives (Sigmoid)

```python
def dgw1(x,w):
    return 1-(np.tanh(x[0]) * np.tanh(x[0]))
def dgw2(x,w):
    return x[0]
def dgw3(x,w):
    return x[1]
def dgw4(x,w):
    return x[2]
def dgw5(x,w):
    return 1

def dgw6(x,w):
    return 1-(np.tanh(x[1]) * np.tanh(x[1]))
def dgw7(x,w):
    return x[0]
def dgw8(x,w):
    return x[1]
def dgw9(x,w):
    return x[2]
def dgw10(x,w):
    return 1

def dgw11(x,w):
    return 1-(np.tanh(x[2]) * np.tanh(x[2]))
def dgw12(x,w):
    return x[0]
def dgw13(x,w):
    return x[1]
def dgw14(x,w):
    return x[2]
def dgw15(x,w):
    return 1

def dgw16(x,w):
    return 1
```

### Jacobian Matrix (Sigmoid)

```python
def gderiv(x, w, N):

    D = np.empty((N, 16))

    for i in range(N):
        D[i, 0] = dgw1(x[i], w)
        D[i, 1] = dgw2(x[i], w)
        D[i, 2] = dgw3(x[i], w)
        D[i, 3] = dgw4(x[i], w)
        D[i, 4] = dgw5(x[i], w)
        D[i, 5] = dgw6(x[i], w)
        D[i, 6] = dgw7(x[i], w)
        D[i, 7] = dgw8(x[i], w)
        D[i, 8] = dgw9(x[i], w)
        D[i, 9] = dgw10(x[i], w)
        D[i, 10] = dgw11(x[i], w)
        D[i, 11] = dgw12(x[i], w)
        D[i, 12] = dgw13(x[i], w)
        D[i, 13] = dgw14(x[i], w)
        D[i, 14] = dgw15(x[i], w)
        D[i, 15] = dgw16(x[i], w)
        
    return D
```

### Levenberg-Marquardt Algorithm (Sigmoid)

```python
def LMsigmoid(x, y, w, lamb, N):
    i = 0
    k2 = gfunction(x, w, N)
    r2 = square_sum_resid(k2, y, N)
    R = np.array([])
    R = np.append(R, r2).astype(float)
    w_1 = w
    
    while(r2 > termination):
        D = gderiv(x, w, N)
        b = D@w - gfunction(x, w, N)
        I = np.identity(16)

        w_1 = (np.linalg.pinv(D.T@D+(lamb*(w.T@w)))@(D.T@b + lamb*(I.T@w)))

        k1 = gfunction(x, w, N)
        k2 = gfunction(x, w_1, N)

        r1 = square_sum_resid(k1, y, N)
        r2 = square_sum_resid(k2, y, N)
        
        R = np.append(R, r2).astype(float)
        
        if (r2 < r1):
            lamb = 0.8*lamb
            w = w_1
        else:
            lamb = 2*lamb
            w = w

        i+=1
        
        if (i == max_iterations):
            break
        
    iterations = np.array(list(range(0, i+1)))
    
    return iterations, R, w_1
    
    
```

## Training the Model

```python
lamb = 0.00001
iterations, residual, sweights = LMsigmoid(data, y, w_init, lamb, 500)
plt.figure(figsize=(12, 8))
plt.plot(iterations, residual, '--bo')
plt.axhline(y=termination, color='r', linestyle='-')
```
